```python
import pandas as pd
import numpy as np
exam = pd.read_csv('exam.csv')
# assign() 함수로 파생변수 추가하기
ex_to = exam.assign(total = exam['math'] + exam['english']
                   + exam['science'])
ex_to.head()
```

```python
# 여러 변수 한번에 만들기
ex_tm = exam.assign(
    total = exam['math']+exam['english']+exam['science'],
    mean = (exam['math']+exam['english']+exam['science'])/3)
ex_tm.head()
```

```python
# assign() 함수에 조건문 넣기
exam.assign(test = np.where(exam['science']>=60,'pass','fail'))\
            .head()
```

```python
exam.assign(total = exam['math']+exam['english']+exam['science']) \
            .sort_values('total').head()
```

```python
# 람다 함수를 이용해 데이터 프레임명 줄여 쓰기
exam.assign(total = lambda x: x['math']+x['science']+x['english']).head()
```

```python
# 앞에서 만든 파생 변수를 이용해 다시 파생 변수를 만들땐 반드시 람다 함수를 사용해야함.
exam.assign(total = exam['math']+exam['english']+exam['science'],
            mean = lambda x: x['total']/3).head()
```
