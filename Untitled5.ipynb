```python
import pandas as pd
import numpy as np
import seaborn as sns

mpg = pd.read_csv('mpg.csv')
# drv 이상치 할당
mpg.loc[[9,13,57,92], 'drv'] = 'k'
# cty 이상치 할당
mpg.loc[[28,42,128,202],'cty'] = [3,4,39,42]

mpg['drv'] = np.where(mpg['drv'].isin(['4','f','r'])
                                      ,mpg['drv'],np.nan)
mpg['drv'].isna().sum()
```

```python
pct25 = mpg['cty'].quantile(.25)
pct75 = mpg['cty'].quantile(.75)
iqr = pct75 - pct25
d = pct25 - (iqr*1.5)
u = pct75 + (iqr*1.5)
sns.boxplot(data=mpg,y='cty')
mpg['cty'] = np.where((mpg['cty'] < d) | (mpg['cty'] > u )
                      ,np.nan, mpg['cty'])
mpg['cty'].isna().sum()
```

```python
sns.boxplot(data=mpg,y='cty')
```

```python
mpg.dropna(subset = 'cty')\
    .groupby('drv')\
    .agg(mean_cty = ('cty','mean'))
```
