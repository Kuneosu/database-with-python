```python
import pandas as pd
mpg = pd.read_csv('mpg.csv')
mpg.head()
```

```python
mpg_new = mpg[['category','cty']]
mpg_new.head()
```

```python
mpg_new.query('category == "suv"')['cty'].mean()
```

```python
mpg_new.query('category == "compact"')['cty'].mean()
```

```python
mpg.sort_values('hwy',ascending = False) \
                .query('manufacturer == "audi"') \
                .head()
```

```python
mpg_copy = mpg.copy()
mpg_copy.head()
```

```python
mpg_copy = mpg_copy.assign(total_y = lambda x: x['cty']+x['hwy'])
mpg_copy.head()
```

```python
mpg_copy = mpg_copy.assign(
        mean_y = lambda x : x['total_y']/2)
mpg_copy.head()
```

```python
mpg_copy.sort_values('mean_y',ascending = False).head(3)
```

```python
mpg.assign(total_y = lambda x : x['cty']+x['hwy'],
          mean_y = lambda x : x['total_y']/2) \
        .sort_values('mean_y',ascending = False).head(3)
```

```python
mpg.groupby(['manufacturer','drv'])\
    .agg(mean_cty = ('cty','mean'))\
    .head()
```

```python
mpg.query('manufacturer == "audi"')\
    .groupby(['drv'])\
    .agg(n = ('drv','count'))
```

```python
mpg.query('manufacturer =="audi"').groupby(['drv'])\
    .agg(n=('drv','count'))
```

```python
mpg.query('manufacturer == "audi"')['drv'].value_counts()
```
