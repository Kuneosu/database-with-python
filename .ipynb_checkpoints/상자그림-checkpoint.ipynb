```python
import pandas as pd
import seaborn as sns

mpg = pd.read_csv('mpg.csv')
sns.boxplot(data = mpg, x = 'drv', y = 'hwy')
```

```python
# compact, subcompact, suv 차종 추출
df_mpg = mpg.query('category in ["compact","subcompact","suv"]')
# 상자 그림 만들기
sns.boxplot(data = df_mpg, x = 'category', y = 'cty')
```
